```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Total_Schools = school_data_complete["school_name"].nunique()
Total_Students = school_data_complete["Student ID"].count()
Total_Budget = school_data["budget"].sum()
Avg_Math = school_data_complete["math_score"].mean()
Avg_reading = school_data_complete["reading_score"].mean()
```

```python
passing_math = school_data_complete.loc[school_data_complete["math_score"]>=70,:]
ttl_passing_math = passing_math["student_name"].count()
pct_passing_math = (ttl_passing_math/Total_Students)*100

passing_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70,:]
ttl_passing_reading = passing_reading["student_name"].count()
pct_passing_reading = (ttl_passing_reading/Total_Students)*100

passing_overall = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]
ttl_passing_overall = passing_overall["student_name"].count()
pct_passing_overall = (ttl_passing_overall/Total_Students)*100
```

```python
Summary_df = pd.DataFrame({"Total Schools": [Total_Schools], "Total Students" : [Total_Students], "Total Budget" : [Total_Budget], "Average Math Score" : [Avg_Math], "Average Reading Score" : [Avg_reading],
                        "% Passing Math" : [pct_passing_math], "% Passing Reading" : [pct_passing_reading], "% Overall Passing" : [pct_passing_overall]})
```

```python
Summary_df["Total Students"] = Summary_df["Total Students"].map("{:,}".format)
Summary_df["Total Budget"] = Summary_df["Total Budget"].map("${:,.2f}".format)
Summary_df
```

```python
Summary_schools = pd.DataFrame({"School Type":type,"Total Students":total_students, "Total School Budget":total_school_budget,"Per Student Budget":per_student_budget, "Average Math Score":ave_math,
                               "Average Reading Score":ave_reading, "% Passing Math":pct_passing_school_math,"% Passing Reading":pct_passing_school_reading,"% Overall Passing":pct_passing_school_overall})

Summary_schools["Total School Budget"] = Summary_schools["Total School Budget"].map("${:,.2f}".format)
Summary_schools["Per Student Budget"] = Summary_schools["Per Student Budget"].map("${:,.2f}".format)
Summary_schools
```

## Math Scores by Grade

```python
math_scores = student_data[["school_name","grade","math_score"]]
math_scores.head()
```

```python
math_9th = math_scores.loc[math_scores["grade"]=="9th",:]
math_9th_score = math_9th[["school_name","math_score"]]
math_9th_group = math_9th_score.groupby(["school_name"])
math_9th_avg = math_9th_group["math_score"].mean()

math_10th = math_scores.loc[math_scores["grade"]=="10th",:]
math_10th_score = math_10th[["school_name","math_score"]]
math_10th_group = math_10th_score.groupby(["school_name"])
math_10th_avg = math_10th_group["math_score"].mean()

math_11th = math_scores.loc[math_scores["grade"]=="11th",:]
math_11th_score = math_11th[["school_name","math_score"]]
math_11th_group = math_11th_score.groupby(["school_name"])
math_11th_avg = math_11th_group["math_score"].mean()

math_12th = math_scores.loc[math_scores["grade"]=="12th",:]
math_12th_score = math_12th[["school_name","math_score"]]
math_12th_group = math_12th_score.groupby(["school_name"])
math_12th_avg = math_12th_group["math_score"].mean()

math_9th_10th = pd.merge(math_9th_avg,math_10th_avg,on = "school_name")
math_11th_12th = pd.merge(math_11th_avg,math_12th_avg,on = "school_name")
math_ttl = pd.merge(math_9th_10th,math_11th_12th,on = "school_name")
math_ttl
```

```python
math_grade = math_ttl.rename(columns={"math_score_x_x":"9th","math_score_y_x":"10th",
                                      "math_score_x_y":"11th","math_score_y_y":"12th"})
math_grade.index.name = None
math_grade
```

## Scores by School Spending

```python
school_group = school.groupby(["school_name"])
type = school_data.set_index(["school_name"])["type"]
total_students = school_group["school_name"].count()
ave_math = school_group["math_score"].mean()
ave_reading = school_group["reading_score"].mean()
total_school_budget = school_group["budget"].mean()
per_student_budget = total_school_budget/total_students

passing_school_math = passing_math.groupby(["school_name"])
ttl_passing_school_math = passing_school_math["student_name"].count()
pct_passing_school_math = (ttl_passing_school_math/total_students)*100

passing_school_reading = passing_reading.groupby(["school_name"])
ttl_passing_school_reading = passing_school_reading["student_name"].count()
pct_passing_school_reading = (ttl_passing_school_reading/total_students)*100

passing_school_overall = passing_overall.groupby(["school_name"])
ttl_passing_school_overall = passing_school_overall["student_name"].count()
pct_passing_school_overall = (ttl_passing_school_overall/total_students)*100

Summary_schools_2 = pd.DataFrame({"Per Student Budget":per_student_budget, "Average Math Score":ave_math,
                                  "Average Reading Score":ave_reading, "% Passing Math":pct_passing_school_math,
                                  "% Passing Reading":pct_passing_school_reading,"% Overall Passing":pct_passing_school_overall})
Summary_schools_2.head()
```
